```python
from ibm_watson import SpeechToTextV1 
import json
from ibm_cloud_sdk_core.authenticators import IAMAuthenticator
```

```python
url_s2t = "https://stream.watsonplatform.net/speech-to-text/api"
```

```python
%store -r iam_apikey_s2t
```

```python
authenticator = IAMAuthenticator(iam_apikey_s2t)
s2t = SpeechToTextV1(authenticator=authenticator)
s2t.set_service_url(url_s2t)
s2t
```

```python
!wget -O AudioTest.mp3  https://speech-translator-audio-test.s3.us-east-2.amazonaws.com/Being+an+Introvert+is+a+Good+Thing.+Crystal+Robello+TEDxStMaryCSSchool.mp3
```

```python
filename='AudioTest.mp3'
```

```python
with open(filename, mode="rb")  as wav:
    response = s2t.recognize(audio=wav, content_type='audio/mp3')
```

```python
response.result
```

```python
from pandas.io.json import json_normalize

json_normalize(response.result['results'],"alternatives")
```

```python
response
```

```python
recognized_text = [i['alternatives'][0]['transcript'] for i in response.result['results']]
recognized_text
```

```python
from ibm_watson import LanguageTranslatorV3
```

```python
url_lt='https://gateway.watsonplatform.net/language-translator/api'
```

```python
%store -r apikey_lt
```

```python
version_lt='2018-05-01'
```

```python
authenticator = IAMAuthenticator(apikey_lt)
language_translator = LanguageTranslatorV3(version=version_lt,authenticator=authenticator)
language_translator.set_service_url(url_lt)
language_translator
```

```python
from pandas.io.json import json_normalize
```

```python
translation_response = language_translator.translate(\
    text=recognized_text, model_id='en-es')
```

```python
translation=translation_response.get_result()
translation
```

```python
spanish_translation = [i['translation'] for i in translation['translations']]
spanish_translation 
```
